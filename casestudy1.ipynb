```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading dataset to python environment

```python
data1=pd.read_csv(r'C:\Users\Admin\Downloads\cars_data.csv')
data1.head()
```

# Check for null values present in the dataset

```python
data1.isnull()
```

# Male vs Female buyers

```python
sns.countplot(data1['Buyer Gender'])
plt.title('Male vs Female buyers')
```

# Top 5 cars based on their sales price

```python
plt.figure(figsize=(20,10))
plt.bar(data1['Make'],data1['Sale Price'])
plt.title('Sale Price distribution',fontsize=20)
plt.xticks(rotation=90)
plt.grid()
plt.xlabel('Make',fontsize=16)
plt.ylabel('Sale price',fontsize=16)
```

# Least 5 cars based on resell price

```python
plt.figure(figsize=(20,10))
plt.bar(data1['Make'],data1['Resell Price'])
plt.title('Resell price distribution',fontsize=20)
plt.xticks(rotation=90)
plt.grid()
plt.xlabel('Make',fontsize=16)
plt.ylabel('Resell Price',fontsize=16)
```
